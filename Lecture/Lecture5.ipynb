# Brute Force Methods

When to use: When input size is small and we are making a prototype/test.

```
function SelSort(A[0 ... n - 1])
 for i <- 0 to n - 2 do 
   min <- i
   for j <- i + 1 to n - 1 do
     if A[j] < A[min] then
       min <- j
   swap A[i] and A[min]
```

**A Sorting Algorithm is**:
- In place: if no additional memory is require, except sometimes a few unit.
- Stable: if it preserves the original order if the sorting keys are equal.
- Input-insensitive: If run times is mostly independent of input properties other than sizes. Suppose a sorting algorithm sorting an array of string and int, insensitive if the sort for both, assuming same size is equal. Sensitive is otherwise.

Selection sort is:
- In place since it doesn't require additional memory to sort the array.
- It is input-insensitive since we don't have any early exit or base-case.

Closest Pair: Find the pair with minimal Euclidean distance.

``` python
min <- inf
FUNC i <- 0 to n - 2 Do
  FOR j <- i + to n - 2 DO 
    d <- sqrt((x_i - x_j)^2 + (y_i - y_j)^2)
	IF d < min THEN
	  min <- d
	  p1 <- i
	  p2 <- j
  RETURN (p1, p2)
```
