## **Growth Rate and Algorithm Efficiency**

- To compare algorithms asymtotically, not the implementation.
- Count Theoretical operations.
- Make assumptions about the input.

``Word RAM Model``:
>- Memory: An infinite array of words/cells each capable of holding an integer
>- Word size: w = 64 bits for current hardware
>- Unit Cost Operation: reading/writing take constant time, as well as arithmetic operations, and bitwise ops on words take constant time.

`Best, Average or Worst Case?`

- **Worst-Case**:Input that maximise running time.
- **Best-case**:Input that minimise running time.
- **Average-case**:Assume inputs follows a distribution, and we computer the expected running time.
- **Amortised**:Consider a set of/sequence of inputs that cause worst-case and average over them.

Sequential Search Example
```
function SeqSearch(A[0...n-1], K)
  i <- 0
  while i < and A[i] != K do
    i <- i + 1
  if i < n then
    return i
  else 
    return -1
```

- Assuming that A[i] are uniformly random over the possible value
- Prob $\mathbb{p}$ that K is in A
- Expected # of comparisons in the assumption that K $\in$ A $$\sum_{i = 0}^{n - 1}(i + 1) \frac{1}{n} = \frac{n+1}{2}$$
- Expected # of comparison (averge case) = average time of K in A + average time of not in A
$$ \frac{n+1}{2} p + n(1 - p)$$
