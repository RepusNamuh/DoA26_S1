## Operation of Stacks

First in first out.

Operations:
>push()\
>peek()\
>pop()\
>is_empty()\
>size()

- The data types of a stack implementation is dependent on the types of data stores:\
That is, if the pointer to the data type is greater than the data stored itself\
then we use dynamic array. If the data type is greater than the pointer to it\
a linked list would be great for simple Stacks operation

- Though if the algorithm require much more push(), pop() operation, then a linked list would be much better do the Constant time pop() push() compared to the dynamic array worse case of O(n)
